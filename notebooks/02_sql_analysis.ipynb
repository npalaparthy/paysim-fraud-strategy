```python
import pandas as pd
import duckdb
```

```python
con = duckdb.connect()

con.execute("""
CREATE OR REPLACE TABLE paysim AS
SELECT *
FROM read_csv_auto('../data/paysim_transactions.csv');
""")
```

```python
con.execute("SELECT COUNT(*) AS total_rows FROM paysim").df()
```

```python
con.execute("""
SELECT *
FROM paysim
LIMIT 5;
""").df()
```

```python
con.execute("""
SELECT type, COUNT(*) AS total_transactions,
                SUM(isFraud) AS fraud_transactions,
          round(SUM(isFraud) * 1.0 / COUNT(*), 4) AS fraud_rate
FROM paysim
GROUP BY type
            order by fraud_rate DESC
""").df()
```
